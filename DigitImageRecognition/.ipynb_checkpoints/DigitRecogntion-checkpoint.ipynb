```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

import warnings
warnings.filterwarnings('ignore')
```

## 2 - Load MNIST Dataset

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = tf.keras.utils.normalize(x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_test, axis = 1)
```

## 3 - Create and Train Sequential Model NN
- Input Layer: Flattened 28x28 pixel image
- 2 hidden layers
- Output layer: Softmax, sum of probabilites equals 1. Pick digit with highest probability
- Compilter: Use adam optimizer, sparse categorical crossentropy loss
- Train using 5 epochs

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
model.add(tf.keras.layers.Dense(units=128, activation="relu"))
model.add(tf.keras.layers.Dense(units=64, activation="relu"))
model.add(tf.keras.layers.Dense(units=10, activation="softmax"))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

```

## 4 - Calculate Loss and Accuracy on test data
- Calculate loss and accuracy for epochs 1 to 4
- As can be seen, model achieves around 97.5% to 98% accuracy for each epoch num

```python
for i in range(4):
    model.fit(x_train, y_train, epochs=i + 1)
    loss, accuracy = model.evaluate(x_test, y_test)
    print(f"loss and accuracy for {i + 1} epochs")
    print(loss)
    print(f"{accuracy} \n")
```

## 5 - Classify Own Writing For Fun
- Used model with epochs=4 to classify my own drawings

```python
image_number = 0
while os.path.isfile(f"digits/digit{image_number}.png"):
    try:
        img = cv2.imread(f"digits/digit{image_number}.png")[:,:,0]
        img = np.invert(np.array([img]))
        prediction = np.argmax(model.predict(img))
        print(f"The image is probably a {prediction}")
        plt.imshow(img[0], cmap=plt.cm.binary)
        plt.show()
    except:
        print("Error")
    finally:
        image_number += 1
        
        
       
```
